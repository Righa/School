```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer,MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation
```

```python
from tensorflow.keras.optimizers import SGD
```

data

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
student_data=pd.read_csv('/content/drive/MyDrive/student_data.csv')
student_data.shape
```

data preparation

```python
student_data.isnull().values.any()
```

```python
student_data.head()
```

```python
y=LabelBinarizer().fit_transform(student_data.Profession)
```

```python
y.shape
```

```python
x=pd.DataFrame(student_data.drop('Profession',axis=1))
```

```python
x.shape
```

```python
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
x
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,train_size=0.7,random_state=1)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
np.savetxt("test_data.csv", X_test, delimiter=",")
```

training and testing

```python
model=Sequential()

model.add(Dense(64,activation='relu',input_dim=23))
model.add(Dropout(0.5))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(6, activation='softmax'))
```

```python
sgd=SGD(learning_rate=0.01,decay=1e-6,momentum=0.9,nesterov=True)
```

```python
model.compile(loss='categorical_crossentropy',optimizer=sgd,metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=500,batch_size=128)
```

```python
model.evaluate(X_test,Y_test,batch_size=128)
```

```python
pred=model.predict(X_test[0:10], batch_size=None, verbose=0, steps=None)
```
